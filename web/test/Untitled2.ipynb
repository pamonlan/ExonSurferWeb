```python
# Plot the results
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
sns.set_style("whitegrid")
sns.set_context("paper")
```

```python
res = pd.read_csv("../test_speed_2.csv")
res.columns
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
sns.lineplot(x="nro_design", y="time", hue="gene", data=res, ax=ax)
ax.set_ylabel("Time (s)")
ax.set_xlabel("Number of primers")
ax.set_title("Time to design primers")
plt.show()
```
